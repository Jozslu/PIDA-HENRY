```python
import pandas as pd 
import numpy as np
import missingno as msno 
import matplotlib.pyplot as plt
import seaborn as sns
```

### Cargando los datasets

```python
data1 = pd.read_csv("DATA/data_principal/Internet_Accesos_Xtecnologia_totalnac.csv", index_col= False, sep= ",", header= 0)
data2 = pd.read_csv("DATA/data_principal/Internet_Accesos_Xtecnologia_xprovincia.csv", index_col= False, sep= ",", header= 0)
data3 = pd.read_csv("DATA/data_principal/Internet_Accesos_xvelocidad_xprovincia_xrangos.csv", index_col= False, sep= ",", header= 0)
data4 = pd.read_csv("DATA/data_principal/Internet_Ingresos.csv", index_col= False, sep= ",", header= 0)
data5 = pd.read_csv("DATA/data_principal/Internet_Penetracion_xprovincia.csv", index_col= False, sep= ",", header= 0)
data6 = pd.read_csv("DATA/data_principal/Internet_distribucion_Accesos_xvelocidad.csv", index_col= False, sep= ",", header= 0)
data7 = pd.read_csv("DATA/data_principal/historico_velocidad_internet_xprovincia.csv", index_col= False, sep= ",", header= 0)
```

### Analizamos cada nuevo Dataframe 

# data1

```python
# Mostrando los primeros datos.
data1.head()
```

```python
# Analizamos la información de sus datos.
data1.info()
```

```python
# Calculamos en número de celdas vacías en cada colummna.
datos_faltantes = data1.isnull().sum().to_frame()
datos_faltantes = datos_faltantes.rename(columns= {0: " Celdas Vacías"})
datos_faltantes
```

```python
# Visualizamos el "gráfico matriz" y observamos algún posible patrón entre los datos faltantes.
msno.matrix(data1, figsize=(10,5), fontsize=11)
```

```python
# Mostrando la forma del Dataframe data1, cantidad de filas y columnas.
print(data1.shape)
```
